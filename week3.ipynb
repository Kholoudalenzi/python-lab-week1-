```python
colour = input("Enter a traffic light colour(red,yellow, green):").strip().lower()

if colour == "":
  print("Error: No colour entered!")
else:
  match colour:
    case "red":
      print("Action: Stop")
    case "yellow":
      print("Action: Get Ready")
    case "green":
      print("Action: Go")
    case _:
      print("Error: Invalid colour entered")

```

```python
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Choose an option (1-4): "))
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

match choice:
    case 1:
        print("Result:", num1 + num2)
    case 2:
        print("Result:", num1 - num2)
    case 3:
        print("Result:", num1 * num2)
    case 4:
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Cannot divide by zero")
    case _:
        print("Invalid choice")
```

```python
score = float(input("Enter student score: "))

if score >= 50:
    status = "Pass"
else:
    status = "Fail"

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
elif score >= 50:
    grade = "E"
else:
    grade = "F"

print("Status:", status)
print("Grade:", grade)

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("Very Good")
    case "C":
        print("Good")
    case "D":
        print("Satisfactory")
    case "E":
        print("Pass")
    case "F":
        print("Better luck next time")
```

```python
total = 0
for i in range(10):
  num = int(input("Enter a number:"))
  if num == 0:
    break
  if num < 0:
    continue
  if num % 2 == 0:
      total += num

print("Final sum =",total)




```

```python
number = int(input("Enter a number: "))

for i in range(1, 13):
    product = number * i

    if product % 3 == 0:
        continue

    if product > 50:
        break

    print(number, "x", i, "=", product)


```

```python
import random

import random

secret_number = random.randint(1, 20)
attempts = 0

while True:
    guess = input("Enter your guess: ")

    if not guess.isdigit():
        continue

    guess = int(guess)

    if guess < 1 or guess > 20:
        continue

    attempts += 1

    if guess == secret_number:
        print("Correct guess!")
        break
```

```python
n = int(input("Enter a number: "))
count = 0

for i in range(1, n + 1):
    if i > 50:
        break

    if i % 8 == 0:
        continue

    if i % 4 == 0:
        count += 1

print(count)
```
